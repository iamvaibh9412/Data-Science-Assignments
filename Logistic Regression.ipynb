# Assignment - Logistic Regression (bank-full.csv)

```python
# Importing the Libraries

import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv("bank-full.csv",sep=";")
df.head()
```

```python
df.sample(10)
```

```python
df.info()
```

```python
df_1= df.drop(['month','day'],axis=1)
```

```python
df_1=pd.get_dummies(data=df_1, columns=['job','marital','education',"default",'housing','loan','contact','poutcome','y'])
df_1
```

```python
df_1=df_1.drop(['y_no'],axis=1)
```

```python
df_1= df_1.rename({'y_yes':'Subscribed'},axis=1)
```

```python
df_1.corr()
```

```python
df_1.shape
```

```python
df_1.info()
```

```python
X = df_1.iloc[:,0:38]
Y = df_1.iloc[:,38]
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
classifier.fit(X,Y)
```

```python
y_pred = classifier.predict(X)
```

```python
y_pred
```

```python
Y_Predicted = pd.DataFrame({'actual':Y, 'Predicted':y_pred})
Y_Predicted
```

```python
from sklearn.metrics import confusion_matrix

confusion_matrix=confusion_matrix(Y,y_pred)
confusion_matrix
```
